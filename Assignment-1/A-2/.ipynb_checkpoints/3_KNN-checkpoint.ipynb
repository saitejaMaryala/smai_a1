```python
def euclidien_distance(x1,x2):
    sum = np.sum((x1-x2)**2)
    return np.sqrt(sum)
```

```python
def Cosine_distance(x,y):
    numerator = np.dot(x,y)

    mod_x = np.sqrt(np.sum(x**2))
    mod_y = np.sqrt(np.sum(y**2))
    
    dist = numerator/(mod_x*mod_y)
    return -dist

# def Cosine_distance(x,y):
#     numerator = torch.dot(x,y)

#     mod_x = torch.sqrt(torch.sum(x**2))
#     mod_y = torch.sqrt(torch.sum(y**2))
    
#     dist = numerator/(mod_x*mod_y)
#     return -dist
```

```python
#vectorized version
def KNN_get_nearest_neighbors(X_train, Y, k=1, dist_metrics="euclidean", batch_size=1000):
    num_samples = Y.shape[0]
    indices_list = []
    
    for i in range(0, num_samples, batch_size):
        batch_Y = Y[i : i + batch_size]
        
        # Compute distances for the batch
        distances = cdist(batch_Y, X_train, metric=dist_metrics)
        
        # Get the indices of k nearest neighbors
        sorted_indices = np.argsort(distances, axis=1)[:, :k]
        indices_list.append(sorted_indices)
    
    # Concatenate all batch results
    indices = np.concatenate(indices_list, axis=0)
    
    return indices

def KNN_get_labels(X_labels, neighbor_indices):
    # Get the corresponding labels for the k nearest neighbors
    top_labels = np.array(X_labels)[neighbor_indices]  # Shape (num_samples, k)
    
    # Find the most frequent label in each row
    predicted_labels = np.apply_along_axis(lambda x: pd.Series(x).mode().iloc[0], axis=1, arr=top_labels)
    
    return predicted_labels

```

```python
def KNN_get_nearestN_text(X_train,Y,k=1,dist_metrics="euclidean"):

    # Compute pairwise distances between Y and X_train
    if dist_metrics == "euclidean":
        distances = cdist(Y, X_train, metric="euclidean")
    else:
        distances = cdist(Y, X_train, metric="cosine")

    # Get the indices of k nearest neighbors
    sorted_indices = np.argsort(distances, axis=1)[:, :k]

    return sorted_indices
    
def KNN_get_labels_text(X_labels,sorted_indices):
    # Get the corresponding labels
    top_labels = np.array(X_labels)[sorted_indices]  # Shape (num_Y, k)
    # Find the most frequent label in each row
    labels = np.apply_along_axis(lambda x: pd.Series(x).mode().iloc[0], axis=1, arr=top_labels)

    return labels
```

```python
def KNN_get_text_to_image_nearest(X_train,Y,k=1,dist_metrics="euclidean"):

    # Compute pairwise distances between Y and X_train
    if dist_metrics == "euclidean":
        distances = cdist(Y, X_train, metric="euclidean")
    else:
        distances = cdist(Y, X_train, metric="cosine")

    # Get the indices of k nearest neighbors
    sorted_indices = np.argsort(distances, axis=1)[:, :k]

    return sorted_indices
    
def KNN_get_text_to_image_labels(X_labels,sorted_indices):
    # Get the corresponding labels
    print(X_labels.shape)
    top_labels = np.array(X_labels)[sorted_indices]  # Shape (num_Y, k)
    # Find the most frequent label in each row
    # labels = np.apply_along_axis(lambda x: pd.Series(x).mode().iloc[0], axis=1, arr=top_labels)
    return top_labels
```

```python
def mean_reciprocal_rank(predictions, ground_truth):
    ranks = []
    for pred, gt in zip(predictions, ground_truth):
        rank_list = np.where(pred == gt)[0]  # Find the index where GT appears
        ranks.append(1 / (rank_list[0] + 1) if len(rank_list) > 0 else 0)
    return np.mean(ranks)

def precision_at_k(predictions, ground_truth, k=100):
    return np.mean([np.sum(pred[:k] == gt) / k for pred, gt in zip(predictions, ground_truth)])

def hit_rate(predictions, ground_truth, k=100):
    return np.mean([1 if gt in pred[:k] else 0 for pred, gt in zip(predictions, ground_truth)])
```

## Locality Sensitive Hashing

```python
def hash_points(vectors,hyper_planes):

    hash_vector_temp = np.array([np.dot(vector, hyper_planes.T) for vector in vectors])
    hash_vector = [hvt > 0 for hvt in hash_vector_temp]

    hash_vector = np.array([i.astype(int) for i in hash_vector])

    return hash_vector
```

```python
def list_buckets(vectors,labels):
    buckets = {}
    for i in range(len(vectors)):
        hash_str = ''.join(vectors[i].astype('str'))
        if hash_str not in buckets.keys():
            buckets[hash_str] = []
        buckets[hash_str].append(int(labels[i]))
    return buckets
```

```python
nbits = 6
d = 512

hyper_planes = np.random.rand(nbits,d) - .5
hyper_planes.shape
```

```python
def find_nearest_bucket(hashed_y, buckets):
    """
    Find the closest matching bucket by flipping one bit at a time.
    """
    hashed_list = list(hashed_y)  # Convert string to list for mutation
    for i in range(len(hashed_list)):
        flipped = hashed_list.copy()
        flipped[i] = '0' if flipped[i] == '1' else '1'  # Flip bit
        new_hash = ''.join(flipped)
        if new_hash in buckets:
            return new_hash  # Return the first match found
        flipped[i] = '0' if flipped[i] == '1' else '1'  # Flip bit
    return None  # No close match found
```

```python


def get_k_nearest_indices(X_train_s, y, plane_indices, k=1, dist_metrics="euclidean"):
    # Compute distances based on the selected metric
    if dist_metrics == "euclidean":
        dists = np.linalg.norm(X_train_s - y, axis=1)
    elif dist_metrics == "cosine":
        dists = distance.cdist([y], X_train_s, metric='cosine')[0]
    else:
        raise ValueError(f"Unsupported distance metric: {dist_metrics}")
    
    # Get indices of the k smallest distances
    k_nearest_indices = np.argsort(dists)[:k]
    
    # Return original dataset indices corresponding to the nearest neighbors
    return [plane_indices[i] for i in k_nearest_indices]


def ANN_get_k_nearest_indices(X_train, Y, X_labels, hyper_planes, k=1, dist_metrics="euclidean"):
    hashed_data = hash_points(X_train, hyper_planes)
    buckets = list_buckets(hashed_data, [i for i in range(len(X_train))])
    
    nearest_indices_list = []
    
    for y in Y:
        hashed_y = hash_points([y], hyper_planes)
        hashed_y = ''.join(hashed_y[0].astype('str'))
        
        if hashed_y not in buckets:
            hashed_y = find_nearest_bucket(hashed_y, buckets)
            if hashed_y is None:
                print(f"Warning: No similar bucket found for {hashed_y}. Skipping.")
                nearest_indices_list.append([])
                continue
        
        plane_indices = buckets[hashed_y]
        X_train_s = np.array([X_train[i] for i in plane_indices])
        
        nearest_indices = get_k_nearest_indices(X_train_s, y, plane_indices, k, dist_metrics)
        nearest_indices_list.append(nearest_indices)
    
    return nearest_indices_list

def ANN_predict_image_to_image(X_train, Y, X_labels, hyper_planes, k=1, dist_metrics="euclidean"):
    nearest_indices_list = ANN_get_k_nearest_indices(X_train, Y, X_labels, hyper_planes, k, dist_metrics)
    
    labels = [[X_labels[i] for i in indices] for indices in nearest_indices_list]
    return labels
```

### Histogram for frequency of points

```python
def hash_points_plt(vectors, hyper_planes):
    hash_vector_temp = np.array([np.dot(vector, hyper_planes.T) for vector in vectors])
    hash_vector = (hash_vector_temp > 0).astype(int)
    return hash_vector

def list_buckets_plt(vectors):
    buckets = {}
    for i in range(len(vectors)):
        hash_str = ''.join(map(str, vectors[i]))  # Convert binary vector to string
        if hash_str not in buckets:
            buckets[hash_str] = 0
        buckets[hash_str] += 1  # Count occurrences in each bucket
    return buckets

# Iterate over different numbers of hyperplanes
for n_bits in [3, 4, 5, 6]:
    hyper_planes = np.random.rand(n_bits, d) - 0.5
    hashed_data = hash_points_plt(data, hyper_planes)
    buckets = list_buckets_plt(hashed_data)
    
    # Sort buckets by size (descending order)
    sorted_buckets = sorted(buckets.items(), key=lambda x: x[1], reverse=True)
    
    # Extract bucket keys (strings) and their sizes
    bucket_keys = [k for k, v in sorted_buckets]
    bucket_sizes = [v for k, v in sorted_buckets]
    
    # Limit to top 20 buckets for better visualization
    top_n = min(30, len(bucket_keys))
    bucket_keys = bucket_keys[:top_n]
    bucket_sizes = bucket_sizes[:top_n]

    # Plot bar chart
    plt.figure(figsize=(12, 5))
    plt.bar(bucket_keys, bucket_sizes, color='blue', alpha=0.7)
    plt.xlabel('Bucket (Binary String)')
    plt.ylabel('Number of Samples')
    plt.title(f'Bucket Distribution (n_bits={n_bits})')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

### IVF

```python
class IVF:
    def __init__(self, num_clusters=10, metric='euclidean'):
        self.num_clusters = num_clusters
        self.metric = metric
        self.centroids = None
        self.inverted_index = {}
        self.labels = None  # Store labels for lookup

    def build_index(self, vectors, labels):
        """Build an IVF index using K-Means clustering."""
        kmeans = KMeans(n_clusters=self.num_clusters, random_state=42, n_init=10)
        cluster_assignments = kmeans.fit_predict(vectors)
        self.centroids = kmeans.cluster_centers_

        self.inverted_index = {i: [] for i in range(self.num_clusters)}
        for idx, cluster_id in enumerate(cluster_assignments):
            self.inverted_index[cluster_id].append(idx)

        self.labels = np.array(labels)  # Store labels for later retrieval

    def search(self, queries, vectors, nprobe=1, top_k=5):
        """Search for nearest neighbors using IVF with nprobe, returning top_k indices."""
        results = []
        for query in queries:
            # Find the closest nprobe clusters
            cluster_distances = cdist([query], self.centroids, metric=self.metric)[0]
            closest_clusters = np.argsort(cluster_distances)[:nprobe]

            # Gather candidates from the closest clusters
            candidate_indices = []
            for cluster in closest_clusters:
                candidate_indices.extend(self.inverted_index.get(cluster, []))

            # Compute exact distances within candidates
            candidate_vectors = vectors[candidate_indices]
            if self.metric == 'euclidean':
                distances = np.linalg.norm(candidate_vectors - query, axis=1)
            elif self.metric == 'cosine':
                distances = 1 - np.dot(candidate_vectors, query) / (np.linalg.norm(candidate_vectors, axis=1) * np.linalg.norm(query))
            else:
                raise ValueError("Unsupported distance metric")

            # Get top_k nearest neighbor indices
            nearest_indices = np.argsort(distances)[:top_k]
            top_k_indices = [candidate_indices[i] for i in nearest_indices]
            results.append(top_k_indices)

        return np.array(results)

    def get_labels(self, indices):
        """Retrieve labels for given indices."""
        return np.array([[self.labels[i] for i in query_indices] for query_indices in indices])
```

```python
start = time.time()
ivf = IVF(num_clusters=10, metric='euclidean')
ivf.build_index(data, data_labels)

indices = ivf.search(test, data, nprobe=3, top_k=5)
predicted_labels = ivf.get_labels(indices)
# print("Predicted labels shape:", predicted_labels)
# print("truth:",test_labels[:100])

mrr = mean_reciprocal_rank(predicted_labels, test_labels)
precision_100 = precision_at_k(predicted_labels, test_labels, k=5)
hit_rate_100 = hit_rate(predicted_labels, test_labels, k=5)
    
# Print Results
print(f"MRR: {mrr:.4f}")
print(f"Precision@100: {precision_100:.4f}")
print(f"Hit Rate@100: {hit_rate_100:.4f}")

end = time.time()
print(f"time took: {end - start:.5f} sec")
```

```python
#To plot the number of points in clusters and avg number of comparisions
class IVF_Inc_plot:
    def __init__(self, num_clusters=10, metric='euclidean'):
        self.num_clusters = num_clusters
        self.metric = metric
        self.centroids = None
        self.inverted_index = {}

    def build_index(self, vectors, labels):
        """Build an IVF index using K-Means clustering."""
        kmeans = KMeans(n_clusters=self.num_clusters, random_state=42, n_init=10)
        cluster_assignments = kmeans.fit_predict(vectors)
        self.centroids = kmeans.cluster_centers_

        self.inverted_index = {i: [] for i in range(self.num_clusters)}
        for idx, cluster_id in enumerate(cluster_assignments):
            self.inverted_index[cluster_id].append(idx)

        self.labels = np.array(labels)  # Store labels for later retrieval
        self.cluster_sizes = np.array([len(self.inverted_index[i]) for i in range(self.num_clusters)])

    def search(self, queries, vectors, nprobe=1, top_k=5):
        """Vectorized search for nearest neighbors using IVF with nprobe, returning top_k indices."""
        cluster_distances = cdist(queries, self.centroids, metric=self.metric)
        closest_clusters = np.argsort(cluster_distances, axis=1)[:, :nprobe]

        candidate_counts = []
        top_k_indices = []

        for query, clusters in zip(queries, closest_clusters):
            candidates = np.concatenate([self.inverted_index.get(cluster, []) for cluster in clusters])
            candidate_counts.append(len(candidates))

            if len(candidates) == 0:
                top_k_indices.append([])
                continue

            candidate_vectors = vectors[candidates]

            # Compute exact distances
            if self.metric == 'euclidean':
                distances = np.linalg.norm(candidate_vectors - query, axis=1)
            elif self.metric == 'cosine':
                distances = 1 - np.dot(candidate_vectors, query) / (np.linalg.norm(candidate_vectors, axis=1) * np.linalg.norm(query))
            else:
                raise ValueError("Unsupported distance metric")

            nearest_indices = np.argsort(distances)[:top_k]
            top_k_indices.append([candidates[i] for i in nearest_indices])

        return np.array(top_k_indices, dtype=object), np.array(candidate_counts)

    def plot_cluster_distribution(self):
        """Plot the number of points in each cluster."""
        plt.figure(figsize=(10, 5))
        plt.bar(range(self.num_clusters), self.cluster_sizes)
        plt.xlabel("Cluster Index")
        plt.ylabel("Number of Points")
        plt.title("Number of Points in Each Cluster")
        plt.show()

    def plot_comparisons_vs_nprobe(self, queries, vectors, max_nprobe=10, top_k=5):
        """Plot average number of comparisons vs. nprobe."""
        nprobes = range(1, max_nprobe + 1)
        avg_comparisons = []

        for nprobe in nprobes:
            _, candidate_counts = self.search(queries, vectors, nprobe=nprobe, top_k=top_k)
            avg_comparisons.append(np.mean(candidate_counts))

        plt.figure(figsize=(10, 5))
        plt.plot(nprobes, avg_comparisons, marker='o')
        plt.xlabel("nprobe")
        plt.ylabel("Average Number of Comparisons per Query")
        plt.title("Average Comparisons vs. nprobe")
        plt.grid()
        plt.show()
```

```python
ivf = IVF_Inc_plot(num_clusters=10)
ivf.build_index(data, data_labels)
# Plot the cluster distribution
ivf.plot_cluster_distribution()
# Plot the average comparisons vs. nprobe
ivf.plot_comparisons_vs_nprobe(test, data, max_nprobe=4)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data sizes
train_size = 50000
test_size = 10000

# KNN comparisons (brute force)
knn_comparisons = np.full(test_size, train_size)  # Every query compares to all points

# IVF comparisons for different nprobe values
nprobe_values = [1, 3, 5, 7]
ivf_comparisons = {nprobe: np.full(test_size, (train_size / 10) * nprobe) for nprobe in nprobe_values}

# LSH comparisons (Fixed 10 buckets, so N/10 comparisons per query)
lsh_comparisons = np.full(test_size, train_size / 10)  # Each query searches only in one bucket

# Plot comparisons
plt.figure(figsize=(10, 6))
plt.plot(np.arange(test_size), knn_comparisons, label="KNN (Brute Force)", linestyle='--', marker='o')

for nprobe, comps in ivf_comparisons.items():
    plt.plot(np.arange(test_size), comps, label=f"IVF (nprobe={nprobe})", linestyle='-', marker='s')

plt.plot(np.arange(test_size), lsh_comparisons, label="LSH (10 Buckets)", linestyle=':', marker='^')

plt.xlabel("Query Index")
plt.ylabel("Comparisons per Query")
plt.title("Comparison of KNN, IVF, and LSH based on Average Comparisons per Query")
plt.legend()
plt.grid()
plt.show()
```
