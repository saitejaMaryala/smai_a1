```python
class Node:
    def __init__(self, threshold=None, feature=None, left=None, right=None, value=None):
        self.threshold = threshold
        self.feature = feature
        self.left = left
        self.right = right
        self.value = value

    def is_leaf(self):
        return self.value is not None

class Decision_Tree:
    def __init__(self, max_depth=10, min_samples=2, n_features=None, criterion="entropy"):
        self.max_depth = max_depth
        self.min_samples = min_samples
        self.n_features = n_features
        self.criterion = criterion  # "entropy" or "gini"
        self.root = None
        # print("Decision tree min_samples:",self.min_samples)

    def fit(self, X, y):
        self.n_features = X.shape[1] if not self.n_features else min(X.shape[1], self.n_features)
        self.root = self._grow_tree(X, y)

    def _grow_tree(self, X, y, depth=0):
        n_samples, present_features = X.shape
        n_labels = len(np.unique(y))

        if depth >= self.max_depth or n_labels == 1 or n_samples <= self.min_samples:
            l_value = self._most_common_label(y)
            return Node(value=l_value)

        feature_idxs = np.random.choice(present_features, self.n_features, replace=False)
        best_feature, best_threshold = self._best_split(X, y, feature_idxs)

        left_idxs, right_idxs = self._split(X[:, best_feature], best_threshold)
        left_node = self._grow_tree(X[left_idxs, :], y[left_idxs], depth + 1)
        right_node = self._grow_tree(X[right_idxs, :], y[right_idxs], depth + 1)
        
        return Node(best_threshold, best_feature, left_node, right_node)

    def _best_split(self, X, y, feature_idxs):
        best_gain = -1
        split_idx, split_threshold = None, None

        for feat_idx in feature_idxs:
            X_column = X[:, feat_idx]
            thresholds = np.unique(X_column)
            
            for thr in thresholds:
                gain = self._gain(y, X_column, thr)
                
                if gain > best_gain:
                    split_idx = feat_idx
                    split_threshold = thr
                    best_gain = gain

        return split_idx, split_threshold

    def _gain(self, y, X_column, threshold):
        parent_impurity = self._impurity(y)
        left_idxs, right_idxs = self._split(X_column, threshold)

        if len(left_idxs) == 0 or len(right_idxs) == 0:
            return 0

        n = len(y)
        l_n, r_n = len(left_idxs), len(right_idxs)
        l_i, r_i = self._impurity(y[left_idxs]), self._impurity(y[right_idxs])

        children_impurity = (l_n / n) * l_i + (r_n / n) * r_i
        
        return parent_impurity - children_impurity

    def _impurity(self, y):
        if self.criterion == "entropy":
            return self._entropy(y)
        elif self.criterion == "gini":
            return self._gini(y)
        else:
            raise ValueError("Invalid criterion. Choose either 'entropy' or 'gini'")

    def _entropy(self, y):
        counts = np.bincount(y)
        probs = counts / len(y)
        return -np.sum([p * np.log2(p) for p in probs if p > 0])

    def _gini(self, y):
        counts = np.bincount(y)
        probs = counts / len(y)
        return 1 - np.sum(probs ** 2)

    def _split(self, X_column, threshold):
        left_idxs = np.argwhere(X_column <= threshold).flatten()
        right_idxs = np.argwhere(X_column > threshold).flatten()
        return left_idxs, right_idxs

    def _most_common_label(self, y):
        counter = Counter(y)
        return counter.most_common(1)[0][0]

    def predict(self, X):
        return np.array([self._traverse(x, self.root) for x in X])

    def _traverse(self, x, node):
        if node.is_leaf():
            return node.value
        if x[node.feature] <= node.threshold:
            return self._traverse(x, node.left)
        return self._traverse(x, node.right)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

class Node:
    def __init__(self, threshold=None, feature=None, left=None, right=None, value=None):
        self.threshold = threshold
        self.feature = feature
        self.left = left
        self.right = right
        self.value = value

    def is_leaf(self):
        return self.value is not None

class Decision_Tree_for_plot:
    def __init__(self, max_depth=10, min_samples=2, n_features=None, criterion="entropy"):
        self.max_depth = max_depth
        self.min_samples = min_samples
        self.n_features = n_features
        self.criterion = criterion
        self.root = None
        self.feature_importances_ = None

    def fit(self, X, y):
        self.n_features = X.shape[1] if not self.n_features else min(X.shape[1], self.n_features)
        self.feature_importances_ = np.zeros(X.shape[1])
        self.root = self._grow_tree(X, y)

    def _grow_tree(self, X, y, depth=0):
        n_samples, present_features = X.shape
        n_labels = len(np.unique(y))

        if depth >= self.max_depth or n_labels == 1 or n_samples < self.min_samples:
            return Node(value=self._most_common_label(y))

        feature_idxs = np.random.choice(present_features, self.n_features, replace=False)
        best_feature, best_threshold, best_gain = self._best_split(X, y, feature_idxs)
        
        if best_feature is None:
            return Node(value=self._most_common_label(y))
        
        self.feature_importances_[best_feature] += best_gain

        left_idxs, right_idxs = self._split(X[:, best_feature], best_threshold)
        left_node = self._grow_tree(X[left_idxs, :], y[left_idxs], depth + 1)
        right_node = self._grow_tree(X[right_idxs, :], y[right_idxs], depth + 1)
        
        return Node(best_threshold, best_feature, left_node, right_node)

    def _best_split(self, X, y, feature_idxs):
        best_gain = -1
        split_idx, split_threshold = None, None

        for feat_idx in feature_idxs:
            X_column = X[:, feat_idx]
            thresholds = np.unique(X_column)
            
            for thr in thresholds:
                gain = self._gain(y, X_column, thr)
                
                if gain > best_gain:
                    split_idx = feat_idx
                    split_threshold = thr
                    best_gain = gain

        return split_idx, split_threshold, best_gain

    def _gain(self, y, X_column, threshold):
        parent_impurity = self._impurity(y)
        left_idxs, right_idxs = self._split(X_column, threshold)

        if len(left_idxs) == 0 or len(right_idxs) == 0:
            return 0

        n = len(y)
        l_n, r_n = len(left_idxs), len(right_idxs)
        l_i, r_i = self._impurity(y[left_idxs]), self._impurity(y[right_idxs])

        children_impurity = (l_n / n) * l_i + (r_n / n) * r_i
        
        return parent_impurity - children_impurity

    def _impurity(self, y):
        if self.criterion == "entropy":
            return self._entropy(y)
        elif self.criterion == "gini":
            return self._gini(y)
        else:
            raise ValueError("Invalid criterion. Choose either 'entropy' or 'gini'")

    def _entropy(self, y):
        counts = np.bincount(y)
        probs = counts / len(y)
        return -np.sum([p * np.log2(p) for p in probs if p > 0])

    def _gini(self, y):
        counts = np.bincount(y)
        probs = counts / len(y)
        return 1 - np.sum(probs ** 2)

    def _split(self, X_column, threshold):
        left_idxs = np.argwhere(X_column <= threshold).flatten()
        right_idxs = np.argwhere(X_column > threshold).flatten()
        return left_idxs, right_idxs

    def _most_common_label(self, y):
        counter = Counter(y)
        return counter.most_common(1)[0][0]

    def predict(self, X):
        return np.array([self._traverse(x, self.root) for x in X])

    def _traverse(self, x, node):
        if node.is_leaf():
            return node.value
        if x[node.feature] <= node.threshold:
            return self._traverse(x, node.left)
        return self._traverse(x, node.right)

    def plot_feature_importance(self, feature_names=None):
        importances = self.feature_importances_ / np.sum(self.feature_importances_)
        indices = np.argsort(importances)[::-1]

        plt.figure(figsize=(10, 6))
        plt.title("Feature Importance")
        plt.bar(range(len(importances)), importances[indices], align="center")
        plt.xticks(range(len(importances)), feature_names if feature_names else indices, rotation=45)
        plt.xlabel("Feature")
        plt.ylabel("Importance")
        plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from collections import Counter

def visualize_tree(tree, feature_names):
    graph = nx.DiGraph()
    pos = {}
    
    def get_tree_width(node, depth):
        """Calculate the width needed at each depth"""
        if node is None:
            return 0
        if node.is_leaf():
            return 1
        return get_tree_width(node.left, depth + 1) + get_tree_width(node.right, depth + 1)
    
    def traverse(node, parent=None, depth=0, x_min=0, x_max=1):
        if node is None:
            return
            
        # Calculate position
        x_pos = (x_min + x_max) / 2
        feature_name = feature_names[node.feature][:4] + "..." if node.feature is not None else ""
        node_label = f"{feature_name}\n<= {node.threshold:.2f}" if not node.is_leaf() else f"Label: {node.value}"
        node_id = id(node)
        
        # Add node and position
        graph.add_node(node_id, label=node_label)
        pos[node_id] = (x_pos * 20, -depth)  # Multiply x_pos by 20 to increase horizontal spacing
        
        if parent is not None:
            graph.add_edge(parent, node_id)
        
        # Calculate positions for children
        if not node.is_leaf():
            # Adjust the spacing between children
            left_width = get_tree_width(node.left, depth + 1)
            right_width = get_tree_width(node.right, depth + 1)
            total_width = left_width + right_width
            
            if total_width > 0:
                left_ratio = left_width / total_width
                split_point = x_min + (x_max - x_min) * left_ratio
                
                # Recursively position children
                if node.left:
                    traverse(node.left, node_id, depth + 1, x_min, split_point)
                if node.right:
                    traverse(node.right, node_id, depth + 1, split_point, x_max)
    
    # Get total width of the tree
    total_width = get_tree_width(tree.root, 0)
    traverse(tree.root)
    
    # Create the visualization
    plt.figure(figsize=(15, 10))  # Increased figure size
    labels = nx.get_node_attributes(graph, 'label')
    
    # Draw the tree with improved visual parameters
    nx.draw(graph, pos, 
            with_labels=True,
            labels=labels,
            node_size=2000,  # Increased node size
            node_color='lightblue',
            font_size=9,
            font_weight='bold',
            edge_color='gray',
            width=2,
            node_shape='o',
            alpha=0.9,
            bbox=dict(facecolor='white', edgecolor='none', alpha=0.7, pad=0.5))
    
    plt.title("Decision Tree Visualization", pad=20, size=14)
    plt.margins(x=0.1)
    plt.axis('off')
    plt.show()
```
